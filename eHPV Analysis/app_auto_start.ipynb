```python
import requests

APP_NAME = "ehpv-waterfall-analysis-app"  # Replace with your actual Databricks App name

# Get notebook context, workspace URL, and API token
ctx = dbutils.notebook.entry_point.getDbutils().notebook().getContext()
host = ctx.apiUrl().get()
token = ctx.apiToken().get()

# Build the start URL
url = f"{host}/api/2.0/apps/{APP_NAME}/start"

# Send the start request
r = requests.post(url, headers={"Authorization": f"Bearer {token}"})

# Output results
print("Status:", r.status_code)
print(r.text)
r.raise_for_status()  # Raises an exception if status code is not 2xx
print("Started app:", APP_NAME)
```
